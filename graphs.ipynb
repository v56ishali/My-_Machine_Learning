```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.barplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.lineplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.boxplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.histplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.violinplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
iris=sns.load_dataset("iris")
correlation_matrix=iris.drop('species',axis=1).corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
iris=sns.load_dataset("iris")
sns.pairplot(iris,hue="species")
plt.title("Pair Plot")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.countplot(x="day",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.jointplot(x="total_bill",y="tip",data=tips)
plt.title("Scatter Plot")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```
